Generar instancias aleatorias

```python
def generarProductos(cantItems, cantProvedores, minStock, maxStock, minDemand, 
                     maxDemand, minminPedido, maxminpedido, minprecio, maxprecio):
    d = {'nombreprod': [],
        'idprod': [], 
    'stockInicial': [],
     'demandaMensual': [],
     'precioUSD': [],
     'pedidoMinimo': [],
    'provedorID': []
    }
    prods = pd.DataFrame(data=d)
    for i in range(1, cantItems+1):
        nom = "prod"+str(i)
        d2 = {'nombreprod': [nom],
            'idprod': [int(i)], 
        'stockInicial': [int(random.randint(minStock, maxStock))],
         'demandaMensual': [int(random.randint(minDemand, maxDemand))],
         'precioUSD': [int(random.randint(minprecio, maxprecio))],
         'pedidoMinimo': [int(random.randint(minminPedido, maxminpedido))],
        'provedorID': [int(random.randint(1, cantProvedores))]
        }
        agregar = pd.DataFrame(data=d2)
        prods = pd.concat([prods, agregar])
    return prods
```

```python
def generarProvedores(cantProvedores, minminmonto, maxminmonto, puertos):
    d3 = {
      'provedor': [],
      'provedorID': [],
      'minimoMontoPedido': [],
      'puertoEnvio': [],
     }
    infoprov = pd.DataFrame(data=d3)

    for i in range(1, cantProvedores+1):
        nom = "prov"+str(i)
        d4 = {
            
            'provedor': [nom],
          'provedorID': [i],
          'minimoMontoPedido': [int(random.randint(minminmonto, maxminmonto))],
          'puertoEnvio': [int(random.randint(1, puertos))],
         }
        agregar = pd.DataFrame(data=d4)

        infoprov = pd.concat([infoprov, agregar])
    return infoprov
```
